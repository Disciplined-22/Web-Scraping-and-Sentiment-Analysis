```python
def count_syllables_for_text(full_text):
    # Initialize an empty dictionary to store word-syllable count pairs
    syllable_count_dict = {}

    # Function to count syllables in a word
    def count_syllables(word):
        # Check for exceptions: words ending with "es" or "ed"
        if word.endswith("es") or word.endswith("ed"):
            return max(1, sum(word.count(vowel) for vowel in "aeiou") - word.count("e"))
        else:
            return max(1, sum(word.count(vowel) for vowel in "aeiou"))

    # Iterate through each word in the tokenized text
    for word in full_text:
        # Count syllables for the word
        syllables = count_syllables(word)
        # Store the word-syllable count pair in the dictionary
        syllable_count_dict[word] = syllables


    return syllable_count_dict

# Example usage:

syllable_counts = count_syllables_for_text(full_text)
total_syllables = sum(syllable_counts.values())
print(total_syllables)

```

```python
def get_words_with_more_than_two_syllables(syllable_count_dict):
    # Initialize a list to store words with more than two syllables
    words_with_more_than_two_syllables = []

    # Iterate through each word-syllable count pair in the dictionary
    for word, syllable_count in syllable_count_dict.items():
        # Check if the syllable count is greater than 2
        if syllable_count > 2:
            # Add the word to the list
            words_with_more_than_two_syllables.append(word)

    complex_word_count = len(words_with_more_than_two_syllables)
    return complex_word_count

print(get_words_with_more_than_two_syllables(syllable_counts))
```

```python
def average_word_length(text):
    # Initialize variables to store the total number of characters and the total number of words
    total_characters = 0
    total_words = len(text)

    # Iterate through each word in the text
    for word in text:
        # Add the length of each word to the total number of characters
        total_characters += len(word)

    # Calculate the average word length
    if total_words > 0:
        average_length = total_characters / total_words
    else:
        average_length = 0

    return average_length

# Example usage:
text = ["This", "is", "a", "sample", "text"]
avg_length = average_word_length(text)
print("Average Word Length:", avg_length)

```
